```python
import pandas as pd
```

```python
def cargar():
  #datos = pd.read_csv('/content/drive/MyDrive/titanic_3.csv', sep=',')
  #datos = pd.read_csv('/content/drive/MyDrive/titanic_3.csv')
  datos = pd.read_excel('/content/drive/MyDrive/titanic.xlsx')
  return datos
```

```python

#datos = pd.read_csv('/content/drive/MyDrive/titanic_3.csv', sep=',', engine='python')
datos = pd.read_excel('/content/drive/MyDrive/titanic.xlsx')

datos.head(3)
```

```python
# Eliminar filas que tienen al menos 1 valor faltante
df=cargar()
print(df.sample(4))
print(df.shape)
print('*'*50)
df = df.dropna()
print(df.sample(4))
print(df.shape)
print('*'*50)
```

```python
# Eliminar por columnas que tienen al menos 1 valor faltante
df=cargar()
print(df.sample(4))
print(df.shape)
print('*'*50)
df = df.dropna(axis=1)
print(df.sample(4))
print(df.shape)
print('*'*50)
```

```python
# Eliminar solo cuando todos sus valores son nulos
df=cargar()
print(df.sample(4))
print(df.shape)
print('*'*50)
df = df.dropna(axis=0, how='all')  ## (axis=1, how='all)
print(df.sample(4))
print(df.shape)
print('*'*50)
```

```python
# Eliminar si tiene menos de 2 faltantes
df=cargar()
print(df.sample(4))
print(df.shape)
print('*'*50)
df = df.dropna(axis=0, thresh=2)  
print(df.sample(4))
print(df.shape)
print('*'*50)
```

```python
# Eliminar por columna específica o dada
df=cargar()
print(df.sample(4))
print(df.shape)
print('*'*50)
df = df.dropna(subset=['Cabin']) 
print(df.sample(4))
print(df.shape)
print('*'*50)
```

```python
# Sustituimos un valor indicado
df=cargar()
print(df.sample(4))
print(df.shape)
print('*'*50)
df.fillna(0, inplace=True)

print(df.sample(4))
print(df.shape)
print('*'*50)
```

```python
# Sustituimos con el promedio
df=cargar()
media=df.mean()
print(df.sample(4))
print(df.shape)
print('*'*50)
df.fillna(media,inplace=True)
print(df.sample(4))
print(df.shape)
print('*'*50)
```

```python
df=cargar()
df.mean()
```

```python
# Sustituimos con valor anterior
df=cargar()
print(df.sample(4))
print(df.shape)
print('*'*50)
df.fillna(method='bfill',inplace=True)
print(df.sample(4))
print(df.shape)
print('*'*50)
```

```python
# Sustituimos con valor posterior
df=cargar()
print(df.sample(4))
print(df.shape)
print('*'*50)
df.fillna(method='ffill',inplace=True)
print(df.sample(4))
print(df.shape)
print('*'*50)
```
