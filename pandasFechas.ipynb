```python
d1 = datetime.datetime.now()
```

```python
print('La fecha de hoy es: ', d1)
```

```python
d2 = d1 + datetime.timedelta(days=2)
print("Pasado ma単ana es ", d2)
```

```python
d3 = d1 + datetime.timedelta(weeks=2)
print("Fecha dos semanas posterior ", d3)
```

```python
# Fechas pasadas
d2 = d1 - datetime.timedelta(days=4)
print("Dia anterior es ", d2)
```
