Muestra un DataFrame con los datos de la siguiente tabla:

```
  Mes     Ingresos     Gastos
  Enero   21500        17000
  Febrero 17600        15400
  Marzo   25000        19200
  Abril   34800        22500
```



````python
datos = {'Mes':['Enero', 'Febrero', 'Marzo', 'Abril'], 
         'Ingresos':[21500, 17600, 25000, 34800], 
         'Gastos':[17000, 15400, 19200, 22500]}
df = pd.DataFrame(datos)
df
````

Escribe una funcion que reciba el DataFrame del ejercicio anterior y devuelva el beneficio total de cada mes (Ingresos - Gastos).


````python
def beneficio(datos):
  datos['beneficios'] = datos.Ingresos - datos.Gastos
  return datos
````

````python
beneficio(df)
````

````python
# Cargamos el dataset
#titanic = pd.read_csv('/content/drive/MyDrive/titanic.csv',index_col=2, delimiter=',')
titanic = pd.read_excel('/content/drive/MyDrive/titanic.xlsx',index_col=2)
titanic.head()
````

````python
# Mostramos dimensiones
print("Dimensiones : ", titanic.shape)
print("Numero de datos: ", titanic.size)
print("Columnas : ", titanic.columns)
print("Filas : ", titanic.index)
print("Tipos de datos : ", titanic.dtypes)
print("Primeras Filas : ", titanic.head(5))
print("Ultimas Filas : ", titanic.tail(5))
````

````python
# Mostrar pasajero 121
titanic[121]
````

````python
# Filas impares
titanic_impares = titanic.iloc[range(0,titanic.shape[0],2)]
titanic_impares.head(15)
````

````python
# Supervivientes
#porcentaje_titanic = titanic.['Survived'].value_counts()/titanic.['Survived'].count()*100
porcentaje_titanic = titanic['Survived'].value_counts()/titanic['Survived'].count()*100
porcentaje_titanic
````

````python
# Segunda clase
#pasajeros_segunda = titanic[titanic['Pclass']==2]['Name'].sort_values()
#pasajeros_segunda
titanic[titanic['Pclass']==1]['Name'].sort_values()
#titanic
````

````python
# Eliminar pasajeros sin edad
titanic_edades = titanic.dropna(subset=['Age'])
titanic_edades
````

````python
# Muertos por clase
#titanic.groupby['Pclass']['Survived'].value_counts()
titanic_por_clase = titanic.groupby['Pclass']['Survived'].value_counts(normalize=True)*100
titanic_por_clase
````

````python
# Edad media por clase
titanic.groupby(['Pclass','Sex'])['Age'].mean()
````

````python
# AÃ±adir columna de menor de edad
titanic['Menor']=titanic['Age']<18
titanic['Menor'].head(15)
````

````python
# Mostrar mayores y menores
titanic.groupby(['Pclass', 'Menor'])['Survived'].value_counts(normalize=True)*100
````
