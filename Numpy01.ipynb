```python
# Otra forma
array_combinado = np.r_[array_1, array_2]
array_combinado
```

```python
# Combinado Horizontalmente
array_combinado = np.hstack((array_1, array_2))
array_combinado
```

```python
# Otra forma
np.column_stack((array_1, array_2))
```

```python
# Otra forma
array_combinado = np.c_[array_1, array_2]
array_combinado
```
