
Construir un array de 4 filas, segun se indica: La primera fila estara formada por los numeros 6,7,9,15,22,34, los elementos de la segunda fila equivalen a la fila anterior multiplicada por 2, la tercera fila equivalen a la fila anterior menos 5 y la cuarta al cuadrado de la tercera fila.



```python
import numpy as np
```

```python
array = np.array([6,7,9,15,22,34])
array
```

```python
array_2 = array * 2
array_3 = array_2 - 5
array_4 = array_3 ** 2
```

```python
array_solucion = np.array([array, array_2, array_3, array_4])
array_solucion
```

```python
pares = array_solucion[array_solucion % 2 == 0]
pares
```
